### Adding text annotation to images

```python
from PIL import Image
import os

# Directory containing the images to be annotated
img_dir = "blueberry_data/train/images"
save_dir = "blueberry_data/train/text_data"
# Loop over each file in the directory
for filename in os.listdir(img_dir):
    # Check that the file is an image (skip any non-image files)
    if filename.endswith(".jpg") or filename.endswith(".png"):
        # Load the image
        img = Image.open(os.path.join(img_dir, filename))

        # Show the image (optional)
        img.show()

        # Prompt the user for annotation text
        annotation = input("Enter annotation text for {}:".format(filename))

        # Save the annotation as a text file with the same name as the image
        with open(os.path.join(save_dir, os.path.splitext(filename)[0] + ".txt"), "w") as f:
            f.write(annotation)
```
